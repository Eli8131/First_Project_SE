```python
lon = np.linspace(-180, 180, 360*2)
lat = np.linspace(-90, 90, 180*2)
```

Individual amplitudes:

```python
count = stats.binned_statistic_2d(ds.lon,
                                ds.lat,
                                ds_diurnal.A1.where(ds.flg_sst2>=4),
                                statistic= 'count',
                                bins = [lon, lat])
```

```python
std = stats.binned_statistic_2d(ds.lon,
                                ds.lat,
                                ds_diurnal.A1.where(ds.flg_sst2>=4),
                                statistic=np.nanstd,
                                bins = [lon, lat])
```

```python
SEMA2 = std.statistic.T / np.sqrt(count.statistic.T)
```

```python
np.savetxt('SEMA1', SEMA1)
```
